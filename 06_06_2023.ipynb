```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./athlete_events.csv')
data.head()
```

```python
pd.set_option('display.max_columns',None)
```

```python
data.drop(columns=['ID','Games'],inplace=True)
```

```python
data.head()
```

```python
data.Medal.unique()
```

```python
data['Gold']=[1 if medal=='Gold' else 0 for medal in data['Medal']]
data['Silver']=[1 if medal=='Silver' else 0 for medal in data['Medal']]
data['Bronze']=[1 if medal=='Bronze' else 0 for medal in data['Medal']]
data.head()
```

```python
c_medals=data.groupby(['Team'])['Gold','Silver','Bronze'].sum()
c_medals
```

```python
c_medals['Total']=c_medals['Gold']+c_medals['Silver']+c_medals['Bronze']
c_medals
```

```python
c_medals.sort_values(by='Total',ascending=False)
```

```python
c_medals.nlargest(1,'Total')
```

```python
no_medal=c_medals[c_medals['Gold']==0]
no_medal
```

```python
list_nm=list(no_medal.index)
no_m=len(list_nm)
no_m
```
