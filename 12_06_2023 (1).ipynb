```python
data['Age']=[(2022-YOB) for YOB in data['YOB']]
data
```

```python
data.drop(columns=['User ID','YOB','Area code'],inplace=True)
data.head()
```

```python
def sort_fn(Location):
    loctn=data[data.Location==Location]
    loctn=loctn.drop(columns='Location')
    return loctn
```

```python
Upper_East_Side_NY=sort_fn('Upper East Side,NY')
Upper_East_Side_NY.head()
```

```python
St_George_NY=sort_fn('St. George,NY')
St_George_NY.head()
```

```python
sns.set_theme(style='darkgrid')
for col in data.columns:
    if data[col].dtypes!='object':
        sns.boxplot(y=data[col],color='red')
        plt.show()
```

```python
for col in data.columns:
    if data[col].dtypes!='object':
        plt.figure(figsize=(6,5))
        sns.histplot(x=data[col],bins=5)
        plt.xticks(rotation=75)
        plt.show()
        
```
