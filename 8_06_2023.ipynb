```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dt=pd.read_csv('./Car_sales (6).csv',parse_dates=['Latest_Launch'])
dt.head()
```

```python
dt.info()
```

```python
dt.describe()
```

```python
dt.isnull().mean()*100
```

```python
dt.sort_values(by='Latest_Launch',inplace=True)
dt.head()
```

```python
dt.reset_index(drop=True,inplace=True)
```

```python
dt
```

```python
dt['year']=dt.Latest_Launch.dt.year
dt['month']=dt.Latest_Launch.dt.month
dt['day']=dt.Latest_Launch.dt.day
```

```python
dt.head()
```

```python
dt.select_dtypes('float64')
```

```python
sns.histplot(data=dt,x='Sales_in_thousands')
plt.show()
```
